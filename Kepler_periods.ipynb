```python
%matplotlib inline
import lightkurve as lk
import matplotlib.pyplot as plt
```

```python
import pandas as pd
import numpy as np
```

```python
search_result = lk.search_lightcurve('KIC 3733346', author='Kepler')
search_result
```

```python
lc = search_result.download()
lc
```

```python
lc.plot()
```

```python
lc_collection = search_result[:5].download_all()
lc_collection
```

```python
fig, ax = plt.subplots(figsize=(20,5))
lc_collection.plot(ax=ax)
```

```python
from astropy.timeseries import LombScargle
frequency, power = LombScargle(lc.time, lc.flux).autopower(minimum_frequency=1/80, maximum_frequency=1/0.01)

print(frequency)
print(power)
```

```python
fig, ax = plt.subplots()
ax.plot(frequency, power)
ax.set_xlabel('Frequency (1/days)')
ax.set_ylabel('Power')
ax.set_title('Lomb-Scargle Periodogram')
plt.show()
```
